```python
generation_design = """
    Hello! I'm your L6σ CoPilot, and I'll help you address your IT help desk challenges using Lean Six Sigma methodology.

## System 1 Response - Quick Innovative Solutions

Based on your situation, here are three innovative approaches to consider:

1. **Hybrid AI-Human Support Model**: Implement a tiered AI solution that handles common issues automatically while seamlessly escalating complex or sensitive matters to human agents, with clear guardrails for data security.

2. **Prompt Engineering Governance Framework**: Develop a robust security-first prompt engineering protocol with regular red-team testing to prevent data leakage while optimizing resolution capabilities.

3. **Knowledge Base Evolution System**: Create a dynamic knowledge repository that continuously learns from successful resolutions while maintaining strict PII/sensitive data filtering.

## System 2 Response - DMAIC Methodology

While these ideas have merit, let's take a methodical approach starting with the Define phase to ensure we address root causes rather than symptoms.

### Define Phase Plan

1. **Project Charter Development**
   - **Problem Statement**: Low first-contact resolution rate (35%) and negative eNPS for IT support, creating inefficiencies and potential security vulnerabilities
   - **Goal Statement**: Increase first-contact resolution to 55%+ and achieve positive eNPS while maintaining zero data breaches
   - **Business Case**: Improved employee productivity, reduced engineering resource diversion, competitive advantage

2. **Scope Definition**
   - IN scope: Help desk processes, AI implementation, security protocols, knowledge management
   - OUT scope: Non-IT service requests, enterprise-wide infrastructure changes

3. **Stakeholder Analysis**
   - Primary: IT leadership, end users, security team
   - Secondary: Executive leadership, engineering teams, compliance/legal
   - Create RACI matrix for key stakeholders

4. **SIPOC Analysis**
   - Suppliers: Ticket submitters, knowledge base contributors
   - Inputs: Support tickets, knowledge articles
   - Process: Current help desk workflow
   - Outputs: Ticket resolutions, escalations
   - Customers: Employees, contractors

5. **Process Mapping**
   - Document current state process flow from ticket creation to resolution
   - Identify decision points and handoffs
   - Map data flows to identify security vulnerabilities

6. **Voice of Customer (VOC)**
   - Conduct targeted surveys on current pain points
   - Interview high-volume ticket submitters
   - Analyze negative eNPS feedback themes

7. **CTQ (Critical to Quality) Tree**
   - Primary branches: Resolution Speed, Resolution Quality, Security/Privacy, User Experience
   - Secondary metrics identification

8. **Project Timeline & Resources**
   - 12-week implementation plan with key milestones
   - Resource requirements and team composition
   - Risk assessment and mitigation plan

9. **Change Management Planning**
   - Identify potential resistance points
   - Develop communication strategy
   - Create training and adoption framework

Remember, improving help desk processes with AI will create organizational change. Some staff may feel threatened by automation, while others may resist new workflows. Plan for stakeholder engagement from the beginning.

Would you like me to elaborate on any specific part of this Define phase plan? Or would you prefer to proceed to the Measure phase to start quantifying the current state in more detail?
"""

system_prompt = f"""
    You will be given a piece of text to evaluate.

    Your task is to rate this text on one specific metric.

    Please read and understand these instructions carefully. Refer to them as needed during your evaluation.

    Evaluation Criteria:

    Define Phase Plan Completeness (1-5) - This metric measures how thoroughly the Define phase plan incorporates key elements of Lean Six Sigma methodology, with a focus on AI-assisted components. A higher score indicates a more comprehensive and technologically advanced plan.

    Evaluation Steps:

    1. Read the provided text carefully.
    2. Check for the presence of the following keywords:
    * Project Charter (AI-Drafted)
    * Problem Statement (Clear)
    * Goal Statement (SMART, AI-Assisted)
    * Scope Definition (In/Out)
    * Change Leadership / OCM
    * Stakeholder Identification & Mapping (AI-Suggested)
    * Voice of Customer (VOC) Analysis (NLP/LLM Summarization)
    * Critical-to-Quality (CTQ) Metrics (AI-Extracted)
    * High-Level Process Map (SIPOC, AI-Templated)
    * Initial KPIs & Metrics (AI-Suggested)
    * Pain Points
    3. Assign a score from 1 to 5 based on the number of keywords present:
        1: 0-2 keywords
        2: 3-5 keywords
        3: 6-8 keywords
        4: 9-10 keywords
        5: All 11 keywords

    Text to Evaluate:

    {generation_design}

    Evaluation Form (scores ONLY):
    Please provide only one number as your output.
    - Define Phase Plan Completeness (1-5):

    Output Format JSON:
        "score": <score>,
        "explanation": "<brief explanation of the score>"
"""

from openai import OpenAI

openai = OpenAI(
    base_url="https://proxy.okareo.com",
    default_headers={"api-key": OKAREO_API_KEY},
    api_key = OPENAI_API_KEY
)

response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_prompt}
    ],
    max_tokens=150,
    temperature=0.7,
)
generated_text = response.choices[0].message.content.strip()
print(f"OpenAI response: {generated_text}")
```
